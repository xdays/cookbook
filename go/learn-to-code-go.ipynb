## Programming Fundamentals

```go
import "fmt"
// bool type
fmt.Println(true, false)

// numberic type
var a1 byte
var a2 uint8
var a3 rune
var a4 int32
fmt.Printf("%T,%T,%T,%T\n", a1, a2, a3, a4)
fmt.Println("use int for integers and float64 for float")

import "runtime"
fmt.Println(runtime.GOOS)
fmt.Println(runtime.GOARCH)

// string type
s := "Hello, 世界"
fmt.Printf("%s\n", s)
fmt.Printf("%q\n", s)
fmt.Printf("%x\n", s)
fmt.Printf("---%x\n", "世")
for i :=0; i < len(s); i++ {
	fmt.Printf("%x ", s[i])
}
fmt.Println()
fmt.Println("--end of string--")

// numeral system

// constant
const b1 = "Hello World"
const b2 int = 42

// iota
const (
    c1 = iota
    c2 = iota
)
fmt.Println(c1, c2)

// bit shifting
x := 4
fmt.Printf("%d\t\t%b\n", x, x)

y := x << 1
fmt.Printf("%d\t\t%b", y, y)
```

## Control Flow

```go
// Control Flow
// for init; condition; post {}
import "fmt"
for a := 1; a < 4; a ++ {
    fmt.Println(a)
}

for b := 1; b < 3; b ++ {
    for c := 1; c < 3; c ++ {
        fmt.Printf("%d --- %d\n", b, c)
    }
}

c := 1
for c < 3 {
    fmt.Printf("c is %d\n", c)
    c++
    fmt.Println("Done!")
}

d := 1
for {

    if d > 3 {
        break
        fmt.Println("Done!")
    }
    d++
    fmt.Printf("d is %d\n", d)
}

// if statement
if true {
    fmt.Println("I am here")
}

if !false {
    fmt.Println("I am still here")
}

if e := 40; e == 42 {
    fmt.Println("e is equal to 42")
} else if e == 40 {
    fmt.Println("e is equal to 40")
} else {
    fmt.Println("e is not equal to 40 or 42")
}

// switch statement
switch {
    case false:
        fmt.Println("this will never be printed")
    case (3==3):
        fmt.Println("true 3")
    default:
        fmt.Println("this default case")
}
// conditional logic operators
fmt.Println(true && true)
fmt.Println(false || true)
```

## Grouping Data

```go
import "fmt"
// array
var x [5]int
fmt.Println(x)

// slice
a := []int{1,2,3,4}
fmt.Println(a)
fmt.Println(len(a))
fmt.Println(len(a))

fmt.Println(a[0])
for i, v := range a {
    fmt.Println(i, v)
}

fmt.Println(a[1:])
fmt.Println(a[1:2])

b := append(a, 5)
fmt.Println(b)
a = append(a, b...)
fmt.Println(a)

c :=append(a[:2], a[4:]...)
fmt.Println(c)

d := make([]int, 10, 100)
fmt.Println(d)
fmt.Println(cap(d))

e1 := []string{"lily", "lilei"}
e2 := []string{"xdays", "ydays"}
e3 := [][]string{e1, e2}
fmt.Println(e3)

// map
m := map[string]int{
    "lily": 15,
    "xdays": 30,
}
fmt.Println(m)
fmt.Println(m["xdays"])
// man, this is awesome!!!
if v, ok := m["ydays"]; ok {
    fmt.Println(v, ok)
}
for k,v := range m {
    fmt.Println(k, v)
}
delete(m, "xdays")
delete(m, "ydays")
fmt.Println(m)
```

## Struct

```go
import "fmt"
// struct

type person struct{
    first string
    last string
    age int
}

p1 := person{
    first: "James",
    last: "Bond",
    age: 30,
}

fmt.Println(p1.first, p1.last, p1.age)

type secretAgent struct{
    person
    ltk bool
}
sa1 := secretAgent {
    person: person{
        first: "James",
        last: "Bond",
        age: 30,
    },
    ltk: true,
}
fmt.Println(sa1.first, sa1.last, sa1.age, sa1.ltk)
```

## Ninja Level 5

```go
// Ninja Level 5
import "fmt"
// E5.1
type person struct{
    first_name string
    last_name string
    flavors []string
}

a1 := person{
    first_name: "xdays",
    last_name: "chen",
    flavors: []string{"hot", "cold"},
}
fmt.Println(a1)
for i,v := range a1.flavors{
    fmt.Println(i, v)
}


// E5.2
persons := map[string]person{
    a1.last_name: a1,
}
fmt.Println(persons)
for _,v := range persons {
    fmt.Println(v.first_name)
    fmt.Println(v.last_name)
    for i,v2 := range v.flavors {
        fmt.Println(i, v2)
    }
}

// E5.3
type vehicle struct{
    color string
    doors int
}
type truck struct{
    vehicle
    fourWheel bool
}
type sedan struct{
    vehicle
    luxury bool
}
b1 := truck{
    vehicle: vehicle{
        color: "red",
        doors: 4,
    },
    fourWheel: true,
}
b2 := sedan{
    vehicle: vehicle{
        color: "blue",
        doors: 5,
    },
    luxury: false,
}
fmt.Println(b1)
fmt.Println(b2)
fmt.Println(b1.doors)
fmt.Println(b2.doors)

// E5.4
c := struct{
    color string
    door int
}{
    color: "red",
    door: 4,
}
fmt.Println(c)
```

## Function

```go
import "fmt"
// function
func foo1() {
    fmt.Println("I am a function")
}
func bar1(s string) (string, bool) {
    return fmt.Sprint("Hello, ", s), true
}
foo1()
v1, v2 := bar1("xdays")
fmt.Println(v1, v2)

// variadic params
func foo2(x ...int) {
    fmt.Println(x)
    fmt.Printf("%T\n", x)
    
    sum := 0
    for _,v := range x {
        sum += v
    }
    fmt.Printf("Sum is: %d\n", sum)
}
foo2(2,3,3,4)

// unfurling slice
a1 := []int{1, 2, 3, 4}
foo2(a1...)

// variadic params must be last param
func foo3(s string, x ...int) {
    fmt.Println(x)
    fmt.Printf("%T\n", x)
}
foo3("xdays", 1, 2, 3)

// defer
func foo4() {
    defer foo1()
    bar1("xdays")
}
foo4()

// methods
// methods is no more than a func attached to type
type person struct{
    first string
    last string
}

func (s person) speak() {
    fmt.Println("I am", s.first, s.last)
}

func test1() {
    p1 := person{
        first: "xdays",
        last: "chen",
    }
    fmt.Println(p1)
    // p1.speak()
}
//test1()

// interfaces & polymorphism
type human interface{
    speak()
}

func humanSpeak(h human) {
    fmt.Println("I call human ", h)
}

func test2() {
    p2 := person{
        first: "xdays",
        last: "chen",
    }
    // humanSpeak(p2)
}
// test2()

// a value can be of more than one type
// hey baby, if you have my method then you're my type
// https://play.golang.org/p/rZH2Efbpot

// anonymous function
func test3() {
    func() {
        fmt.Println("Anonymous function here")
    }()
}
test3()

// function expression
f := func() {
    fmt.Println("My first func epxression")
}
f()

// return a function
func foo5() (func() int) {
    return func() int {
        return 35
    }
}
test4 := foo5()
fmt.Printf("%T\n", test4)
fmt.Println("test4 function result:",test4())

// callback
func sum(x ...int) int {
    t := 0
    for _,v := range x {
        t += v
    }
    return t
}

func even(f func(x ...int) int, y ...int) int {
    var y1 []int
    for _, v := range y {
        if v%2 == 0 {
            y1 = append(y1, v)
        }
    }
    fmt.Println("even numbers:", y1)
    return f(y1...)
}
fmt.Println(even(sum, 1,2,3,4,5))
```
