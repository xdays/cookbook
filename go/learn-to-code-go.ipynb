## Programming Fundamentals

```go
import "fmt"
// bool type
fmt.Println(true, false)

// numberic type
var a1 byte
var a2 uint8
var a3 rune
var a4 int32
fmt.Printf("%T,%T,%T,%T\n", a1, a2, a3, a4)
fmt.Println("use int for integers and float64 for float")

import "runtime"
fmt.Println(runtime.GOOS)
fmt.Println(runtime.GOARCH)

// string type
s := "Hello, 世界"
fmt.Printf("%s\n", s)
fmt.Printf("%q\n", s)
fmt.Printf("%x\n", s)
fmt.Printf("---%x\n", "世")
for i :=0; i < len(s); i++ {
	fmt.Printf("%x ", s[i])
}
fmt.Println()
fmt.Println("--end of string--")

// numeral system

// constant
const b1 = "Hello World"
const b2 int = 42

// iota
const (
    c1 = iota
    c2 = iota
)
fmt.Println(c1, c2)

// bit shifting
x := 4
fmt.Printf("%d\t\t%b\n", x, x)

y := x << 1
fmt.Printf("%d\t\t%b", y, y)
```

## Control Flow

```go
// Control Flow
// for init; condition; post {}
import "fmt"
for a := 1; a < 4; a ++ {
    fmt.Println(a)
}

for b := 1; b < 3; b ++ {
    for c := 1; c < 3; c ++ {
        fmt.Printf("%d --- %d\n", b, c)
    }
}

c := 1
for c < 3 {
    fmt.Printf("c is %d\n", c)
    c++
    fmt.Println("Done!")
}

d := 1
for {

    if d > 3 {
        break
        fmt.Println("Done!")
    }
    d++
    fmt.Printf("d is %d\n", d)
}

// if statement
if true {
    fmt.Println("I am here")
}

if !false {
    fmt.Println("I am still here")
}

if e := 40; e == 42 {
    fmt.Println("e is equal to 42")
} else if e == 40 {
    fmt.Println("e is equal to 40")
} else {
    fmt.Println("e is not equal to 40 or 42")
}

// switch statement
switch {
    case false:
        fmt.Println("this will never be printed")
    case (3==3):
        fmt.Println("true 3")
    default:
        fmt.Println("this default case")
}
// conditional logic operators
fmt.Println(true && true)
fmt.Println(false || true)
```

## Ninja Level 3

```go
// Ninja Level 3
import "fmt"
// E3.1
for i:=1; i<10; i++ {
    fmt.Println(i)
}

// E3.2
for i:=65; i<=70; i++ {
    fmt.Println(i)
    for j := 1; j<=3; j++ {
        fmt.Printf("\t%#U\n", i)
    }
    
}

// E3.3
bd := 1999
for bd <= 2019 {
    fmt.Println(bd)
    bd++
}

// E3.4
bd2 := 1999
for {
    if bd2 > 2019 {
        break
    }
    fmt.Println(bd2)
    bd2++
}

// E3.5
for i := 10; i <= 20; i++ {
	fmt.Printf("When %v is divided by 4, the remainder (aka modulus) is %v\n", i, i%4)
}

// E3.6
a1 := 3
if a1 == 3 {
    fmt.Println("a1 is equal to 3")
}

// E3.7
a2 := 3
if a1 == 3 {
    fmt.Println("a1 is equal to 3")
} else if a1 == 4 {
    fmt.Println("a1 is equal to 4")
} else {
    fmt.Println("a1 is not equal to 3 or 4")
}

// E3.8
switch {
    case false: {
        fmt.Println("never goes here")
    }
    case (1==1): {
        fmt.Println("1 is equal to 1")
    }
}

// E3.9
favSport := "pingpong"
switch favSport {
    case "basketball": {
        fmt.Println("I like basketball")
    }
    default: {
        fmt.Println("I like pingpong")
    }
}

// E3.10
fmt.Println(true && true)
fmt.Println(true && false)
fmt.Println(true || true)
fmt.Println(true || false)
fmt.Println(!true)
```

## Grouping Data

```go
import "fmt"
// array
var x [5]int
fmt.Println(x)

// slice
a := []int{1,2,3,4}
fmt.Println(a)
fmt.Println(len(a))
fmt.Println(len(a))

fmt.Println(a[0])
for i, v := range a {
    fmt.Println(i, v)
}

fmt.Println(a[1:])
fmt.Println(a[1:2])

b := append(a, 5)
fmt.Println(b)
a = append(a, b...)
fmt.Println(a)

c :=append(a[:2], a[4:]...)
fmt.Println(c)

d := make([]int, 10, 100)
fmt.Println(d)
fmt.Println(cap(d))

e1 := []string{"lily", "lilei"}
e2 := []string{"xdays", "ydays"}
e3 := [][]string{e1, e2}
fmt.Println(e3)

// map
m := map[string]int{
    "lily": 15,
    "xdays": 30,
}
fmt.Println(m)
fmt.Println(m["xdays"])
// man, this is awesome!!!
if v, ok := m["ydays"]; ok {
    fmt.Println(v, ok)
}
for k,v := range m {
    fmt.Println(k, v)
}
delete(m, "xdays")
delete(m, "ydays")
fmt.Println(m)
```
