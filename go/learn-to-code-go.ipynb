## Control Flow

```go
// Control Flow
// for init; condition; post {}
import "fmt"
for a := 1; a < 4; a ++ {
    fmt.Println(a)
}

for b := 1; b < 3; b ++ {
    for c := 1; c < 3; c ++ {
        fmt.Printf("%d --- %d\n", b, c)
    }
}

c := 1
for c < 3 {
    fmt.Printf("c is %d\n", c)
    c++
    fmt.Println("Done!")
}

d := 1
for {

    if d > 3 {
        break
        fmt.Println("Done!")
    }
    d++
    fmt.Printf("d is %d\n", d)
}

// if statement
if true {
    fmt.Println("I am here")
}

if !false {
    fmt.Println("I am still here")
}

if e := 40; e == 42 {
    fmt.Println("e is equal to 42")
} else if e == 40 {
    fmt.Println("e is equal to 40")
} else {
    fmt.Println("e is not equal to 40 or 42")
}

// switch statement
switch {
    case false:
        fmt.Println("this will never be printed")
    case (3==3):
        fmt.Println("true 3")
    default:
        fmt.Println("this default case")
}
// conditional logic operators
fmt.Println(true && true)
fmt.Println(false || true)
```

## Function

```go
import "fmt"
// function
func foo1() {
    fmt.Println("I am a function")
}
func bar1(s string) (string, bool) {
    return fmt.Sprint("Hello, ", s), true
}
foo1()
v1, v2 := bar1("xdays")
fmt.Println(v1, v2)

// variadic params
func foo2(x ...int) {
    fmt.Println(x)
    fmt.Printf("%T\n", x)
    
    sum := 0
    for _,v := range x {
        sum += v
    }
    fmt.Printf("Sum is: %d\n", sum)
}
foo2(2,3,3,4)

// unfurling slice
a1 := []int{1, 2, 3, 4}
foo2(a1...)

// variadic params must be last param
func foo3(s string, x ...int) {
    fmt.Println(x)
    fmt.Printf("%T\n", x)
}
foo3("xdays", 1, 2, 3)

// defer
func foo4() {
    defer foo1()
    bar1("xdays")
}
foo4()

// methods
// methods is no more than a func attached to type
type person struct{
    first string
    last string
}

func (s person) speak() {
    fmt.Println("I am speaking from", s.first, s.last)
}

func test1() {
    p1 := person{
        first: "xdays",
        last: "chen",
    }
    fmt.Println(p1)
    p1.speak()
}
test1()

// interfaces & polymorphism
type human interface{
    speak()
}

func humanSpeak(h human) {
    fmt.Println("I call human ", h)
}

func test2() {
    p2 := person{
        first: "xdays",
        last: "chen",
    }
    humanSpeak(p2)
}
test2()

// a value can be of more than one type
// hey baby, if you have my method then you're my type
// https://play.golang.org/p/rZH2Efbpot

// anonymous function
func test3() {
    func() {
        fmt.Println("Anonymous function here")
    }()
}
test3()

// function expression
f := func() {
    fmt.Println("My first func epxression")
}
f()

// return a function
func foo5() (func() int) {
    return func() int {
        return 35
    }
}
test4 := foo5()
fmt.Printf("%T\n", test4)
fmt.Println("test4 function result:",test4())

// callback
func sum(x ...int) int {
    t := 0
    for _,v := range x {
        t += v
    }
    return t
}

func even(f func(x ...int) int, y ...int) int {
    var y1 []int
    for _, v := range y {
        if v%2 == 0 {
            y1 = append(y1, v)
        }
    }
    fmt.Println("even numbers:", y1)
    return f(y1...)
}
fmt.Println(even(sum, 1,2,3,4,5))

// closure
// package level scope
// function level scope
// block level scope
func incrementor() func() int {
    var x int
    return func() int {
        x++
        return x
    }
}

a := incrementor()
b := incrementor()
fmt.Println(a())
fmt.Println(a())
fmt.Println(b())

// recursion
func factorial1(n int) int {
    if n == 0 {
        return 1
    }
    return n * factorial1(n-1)
}

n1 := factorial1(4)
fmt.Println("Result by recursion:", n1)

func factorial2(n int) int {
    s := 1
    for ; n>0; n-- {
        s *= n
    }
    return s
}
n2 := factorial2(4)
fmt.Println("Result by loop:", n2)
```

## Pointer

```go
import "fmt"
// Pointer
// 1. to avoid pass around lots of data
// 2. change data in place
// Everything in Go is pass by value

a1 := 42
fmt.Println(a1)
fmt.Println(&a1)
fmt.Printf("a1 is a %T\n", &a1)

var b *int = &a1
fmt.Println(b)
fmt.Println(*&a1)

*b = 43 // change in place, not new address
fmt.Println(*b)
fmt.Println(a1)

// Pointer Usage
func change(x *int) {
    *x = 43
}
a2 := 0
fmt.Println("a2 addr before", &a2)
fmt.Println("a2 after", a2)
change(&a2)
fmt.Println("a2 addr before", &a2)
fmt.Println("a2 after", a2)

// Method Sets
// non-pointer receiver works with value that are pointers and non-pointers
// pointer receive works with value that are pointers
```

```go
import "fmt"
import "encoding/json"

// json
type person struct{
    First string
    Last string
    Age int
}

p1 := person{
    First: "xdays",
    Last: "chen",
    Age: 30,
}

people := []person{p1,}

fmt.Println(people)
bs, err := json.Marshal(people)
if err != nil {
    fmt.Println(err)
}
fmt.Printf("%T\n", bs)
fmt.Println(string(bs))

var ps []person
err2 := json.Unmarshal(bs, &ps)
if err != nil {
    fmt.Println(err2)
}
fmt.Println("All data:", ps)

// Writer
import "io"
import "os"
fmt.Fprintln(os.Stdout, "hello os!")
io.WriteString(os.Stdout, "Hello io!\n")

// Sort
import "sort"
x1 := []int{1,3,2,6,4}
x2 := []string{"a", "c", "b"}
fmt.Println(x1)
sort.Ints(x1)
fmt.Println(x1)
fmt.Println(x2)
sort.Strings(x2)
fmt.Println(x2)

// Custom Sort
import "sort"
type Person struct {
	First string
	Age   int
}

type ByAge []Person
func (a ByAge) Len() int { return len(a) }
func (a ByAge) Swap(i, j int) { a[i], a[j] = a[j], a[i] }
func (a ByAge) Less(i, j int) bool { return a[i].Age < a[j].Age }

type ByName []Person
func (a ByName) Len() int { return len(a) }
func (a ByName) Swap(i, j int) { a[i], a[j] = a[j], a[i] }
func (a ByName) Less(i, j int) bool { return a[i].First < a[j].First }

p1 := Person{"James", 32}
p2 := Person{"Moneypenny", 27}
p3 := Person{"Q", 64}
p4 := Person{"M", 56}

people := []Person{p1, p2, p3, p4}
fmt.Println(people)
sort.Sort(ByName(people))
fmt.Println(people)
sort.Sort(ByAge(people))
fmt.Println(people)

// Bcrypt
import "golang.org/x/crypto/bcrypt"
s := `password123`
bs, err := bcrypt.GenerateFromPassword([]byte(s), bcrypt.MinCost)
if err != nil {
    fmt.Println(err)
}
fmt.Println(s)
fmt.Println(bs)

loginPwd := `password123`
err := bcrypt.CompareHashAndPassword(bs, []byte(loginPwd))
if err != nil {
    fmt.Println("YOU CAN'T LOGIN")
}
fmt.Println("You're logged in")
```
